```sql
--1.feladat Group by cube
SELECT 
    S.SZALLAS_ID,
    S.SZALLAS_NEV,
    S.HELY AS SZALLAS_HELYE,
    COUNT(F.FOGLALAS_PK) AS FOGLALASOK_SZAMA,
    V.NEV AS VENDEG_NEVE
FROM 
    SZALLASHELY S
JOIN 
    SZOBA SZ ON S.SZALLAS_ID = SZ.SZALLAS_FK
JOIN 
    FOGLALAS F ON SZ.SZOBA_ID = F.SZOBA_FK
JOIN 
    VENDEG V ON F.UGYFEL_FK = V.USERNEV
GROUP BY 
    CUBE (S.SZALLAS_ID, S.SZALLAS_NEV, S.HELY, V.NEV);

```

```sql
--Group by Groupingsets
SELECT 
    SZ.SZALLAS_ID,
    SZ.SZALLAS_NEV,
    SZO.SZOBA_SZAMA,
    COUNT(FO.FOGLALAS_PK) AS FOGLALASOK_SZAMA,
    SUM(FO.FELNOTT_SZAM) AS OSSZES_FELNOTT,
    SUM(FO.GYERMEK_SZAM) AS OSSZES_GYERMEK
FROM 
    SZALLASHELY SZ
JOIN 
    SZOBA SZO ON SZ.SZALLAS_ID = SZO.SZALLAS_FK
LEFT JOIN 
    FOGLALAS FO ON SZO.SZOBA_ID = FO.SZOBA_FK
GROUP BY 
    GROUPING SETS ((SZ.SZALLAS_ID, SZ.SZALLAS_NEV, SZO.SZOBA_SZAMA), ())
ORDER BY 
    SZ.SZALLAS_ID, SZO.SZOBA_SZAMA;
```

```sql
-- groupby rollup
SELECT
    CASE WHEN GROUPING(S.SZALLAS_NEV) = 1 THEN 'ÖSSZES SZÁLLÁS' ELSE S.SZALLAS_NEV END AS SZALLAS_NEV,
    F.METTOL,
    F.MEDDIG,
    SUM(F.FELNOTT_SZAM) AS OSSZ_FELNOTT,
    SUM(F.GYERMEK_SZAM) AS OSSZ_GYERMEK
FROM
    SZALLASHELY S
JOIN
    SZOBA Z ON S.SZALLAS_ID = Z.SZALLAS_FK
JOIN
    FOGLALAS F ON Z.SZOBA_ID = F.SZOBA_FK
GROUP BY ROLLUP (S.SZALLAS_NEV, F.METTOL, F.MEDDIG)
ORDER BY
    S.SZALLAS_NEV, F.METTOL, F.MEDDIG;
```

```sql
--2.feladat
SELECT 
    SZALLAS_ID,
    SZALLAS_NEV,
    CSILLAGOK_SZAMA,
    ROW_NUMBER() OVER (ORDER BY CSILLAGOK_SZAMA DESC) AS SORSZAM_CSILLAGOK_SZERINT
FROM 
    SZALLASHELY;
```

```sql
SELECT 
    SZOBA_ID,
    SZOBA_SZAMA,
    SZALLAS_NEV,
    NTILE(3) OVER (ORDER BY FEROHELY DESC) AS FEROHELY_CSOPORT
FROM 
    SZOBA
JOIN 
    SZALLASHELY ON SZOBA.SZALLAS_FK = SZALLASHELY.SZALLAS_ID;
```

```sql
--3.feladat 
CREATE PROCEDURE VendegFelhasznalo (
    @GUEST_USERNAME NVARCHAR(20)
)
AS
BEGIN
    SELECT *
    FROM FOGLALAS
    WHERE UGYFEL_FK = @GUEST_USERNAME;
END;
```
